```python
from game.game import simulate, Edge, Strategy, VertexWeight, Graph, Game, Utility, PossibleStrategies
```

### Make a graph to construct a game
Form a graph with edges, strategies, and weights.
- v: vertex
- n: number of vertices in a graph
- Edges: n by n matrix
- Strategies: each v's strategy float number
- Weights: each v's weight (you can ignore it in this game)

```python
# Find "Edge" class methods to get edges
init_edges = Edge.ws_model(n_vertices=10, k_nearest=2, rewire_prob=0.4)

# Find "Strategy" class methods to get strategies
init_strategies = Strategy.arange(n_vertices=len(init_edges))

# Find "VertexWeight" class methods to get weights
init_weights = VertexWeight.same(n_vertices=len(init_edges))

# Use the above three components to form a graph
g = Graph(strategies=init_strategies, edges=init_edges, weights=init_weights)

# Use a graph and a utility function to form a game
# Find "Utility" class methods to get callable utility functions
m = Game(graph=g, utility_func=Utility.maximal_matching_1)
```

### Run a turn
A vertices is like a player in a game.
Find all vertices which have other choice of strategy to improve its utility the most.
Randomly find one of them to switch its strategy.

```python
# Run a move
(is_terminated, have_better_choices, g, selected_vx,
 better_choices) = Game.run_a_time(graph=g,
                                   utility_func=Utility.maximal_matching_2,
                                   strategy_list=PossibleStrategies.all_vertices(g),
                                   opt={})
```

### Plot the current graph and check it

```python
# Plot the graph
#     - Black lines indicate if two vertices are connected with each other.
#     - Orange arrow: choose what vertex
#     - "s": selected strategy
#     - Red lines: a matching
#     - Filled with light-blue: the vertex is already in a matching.
g.plot_graph_maximal_matching(show=True,
                              picked_vertex=selected_vx,
                              have_better_choice=have_better_choices)

# Calculate the total number of matched pair in the graph
print(f'Number of matched pairs: {g.n_matching()}')

# Check if its a maximal matching
# If failed, check the utility function to fix it
print(f'Pass the check? {Game.check_real_maximal_matching(g)}')
```

### Simulate and test different utility functions
Test each utility function with different rewiring probability, 100 games per combination.
The result will be plotted.
